```python
import networkx as nx
import matplotlib.pyplot as plt


def node_subst_cost(n1, n2):
    return 1 if n1 != n2 else 0


def node_del_cost(n):
    return 1


def node_ins_cost(n):
    return 1


def edge_subst_cost(e1, e2):
    return 1 if e1 != e2 else 0


def edge_del_cost(e):
    return 1


def edge_ins_cost(e):
    return 1


def test_graph_edit_distance():
    # Create two sample graphs
    G1 = nx.Graph()
    G1.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

    G2 = nx.Graph()
    G2.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)])

    # Calculate the graph edit distance
    ged = nx.graph_edit_distance(
        G1,
        G2,
        node_subst_cost=node_subst_cost,
        node_del_cost=node_del_cost,
        node_ins_cost=node_ins_cost,
        edge_subst_cost=edge_subst_cost,
        edge_del_cost=edge_del_cost,
        edge_ins_cost=edge_ins_cost,
    )

    # Identify node and edge operations
    nodes_G1 = set(G1.nodes)
    nodes_G2 = set(G2.nodes)
    edges_G1 = set(G1.edges)
    edges_G2 = set(G2.edges)

    inserted_nodes = nodes_G2 - nodes_G1
    deleted_nodes = nodes_G1 - nodes_G2
    substituted_nodes = (
        nodes_G1 & nodes_G2
    )  # For simplicity, assume no substitution if nodes are the same

    inserted_edges = edges_G2 - edges_G1
    deleted_edges = edges_G1 - edges_G2
    substituted_edges = (
        edges_G1 & edges_G2
    )  # For simplicity, assume no substitution if edges are the same

    # Calculate costs
    node_insertion_cost = sum(node_ins_cost(n) for n in inserted_nodes)
    node_deletion_cost = sum(node_del_cost(n) for n in deleted_nodes)
    node_substitution_cost = sum(
        node_subst_cost(n1, n2) for n1, n2 in zip(substituted_nodes, substituted_nodes)
    )

    edge_insertion_cost = sum(edge_ins_cost(e) for e in inserted_edges)
    edge_deletion_cost = sum(edge_del_cost(e) for e in deleted_edges)
    edge_substitution_cost = sum(
        edge_subst_cost(e1, e2) for e1, e2 in zip(substituted_edges, substituted_edges)
    )

    total_node_cost = node_insertion_cost + node_deletion_cost + node_substitution_cost
    total_edge_cost = edge_insertion_cost + edge_deletion_cost + edge_substitution_cost
    total_cost = total_node_cost + total_edge_cost

    print(f"Graph Edit Distance: {ged}")
    print(f"Inserted Nodes: {inserted_nodes}")
    print(f"Deleted Nodes: {deleted_nodes}")
    print(f"Substituted Nodes: {substituted_nodes}")
    print(f"Inserted Edges: {inserted_edges}")
    print(f"Deleted Edges: {deleted_edges}")
    print(f"Substituted Edges: {substituted_edges}")
    print("\nCosts:")
    print(f"Node Insertion Cost: {node_insertion_cost}")
    print(f"Node Deletion Cost: {node_deletion_cost}")
    print(f"Node Substitution Cost: {node_substitution_cost}")
    print(f"Edge Insertion Cost: {edge_insertion_cost}")
    print(f"Edge Deletion Cost: {edge_deletion_cost}")
    print(f"Edge Substitution Cost: {edge_substitution_cost}")
    print(f"Total Node Cost: {total_node_cost}")
    print(f"Total Edge Cost: {total_edge_cost}")
    print(f"Total Cost: {total_cost}")

    # Visualize the graphs
    plt.figure(figsize=(10, 5))

    plt.subplot(121)
    nx.draw(G1, with_labels=True, node_color="lightblue", node_size=500, font_size=16)
    plt.title("Graph 1")

    plt.subplot(122)
    nx.draw(G2, with_labels=True, node_color="lightgreen", node_size=500, font_size=16)
    plt.title("Graph 2")

    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    test_graph_edit_distance()
```
